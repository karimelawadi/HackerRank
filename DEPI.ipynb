List Commander

```python
if __name__ == '__main__':
    
    '''Num List => list that store numbers'''
    Num_list = []

    ''''input number of Commands '''
    n = int(input())  
    
    for Num_Command  in range(n):
        '''input Commands and numbers'''
        action = input().split()

        '''Sart Of Action '''  
        if action[0] == "append":
            Num_list.append(int(action[1]))
        elif action[0] == "insert":
            Num_list.insert(int(action[1]), int(action[2]))
        elif action[0] == "remove":
            Num_list.remove(int(action[1]))
        elif action[0] == "print":
            print(Num_list)
        elif action[0] == "sort":
            Num_list.sort()
        elif action[0] == "pop":
            Num_list.pop()
        elif action[0] == "reverse":
            Num_list.reverse()


```
